```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo dos gr√°ficos
plt.style.use("seaborn-v0_8")
```

```python
df = pd.read_excel("dados_voo.xlsx")

df.head()
```

```python
df["atraso_chegada"] = df["arr_delay"] > 0
```

```python
df[["arr_delay", "atraso_chegada"]].head(10)
```

```python
tabela_companhias = (
    df
    .groupby("carrier")
    .agg(
        total_voos=("carrier", "count"),
        qtd_voos_atrasados=("atraso_chegada", "sum"),
        atraso_medio_min=("arr_delay", "mean")
    )
    .reset_index()
)

tabela_companhias["percentual_atrasos"] = (
    tabela_companhias["qtd_voos_atrasados"] /
    tabela_companhias["total_voos"] * 100
)

tabela_companhias = tabela_companhias.sort_values(
    by="percentual_atrasos",
    ascending=False
)

tabela_companhias
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    data=tabela_companhias,
    x="carrier",
    y="percentual_atrasos"
)

plt.title("Percentual de voos com atraso por companhia a√©rea")
plt.xlabel("Companhia A√©rea")
plt.ylabel("Percentual de Atrasos (%)")

plt.tight_layout()
plt.show()
```

```python
tabela_rotas = (
    df
    .groupby(["origin", "dest"])
    .agg(
        total_voos=("carrier", "count"),
        atraso_medio_min=("arr_delay", "mean"),
        percentual_atrasos=("atraso_chegada", "mean")
    )
    .reset_index()
)

# Converter percentual para %
tabela_rotas["percentual_atrasos"] = tabela_rotas["percentual_atrasos"] * 100

# Filtrar apenas rotas com volume significativo
tabela_rotas = tabela_rotas[tabela_rotas["total_voos"] >= 100]

# Top 10 rotas com MAIOR atraso m√©dio
top_10_piores_rotas = tabela_rotas.sort_values(
    by="atraso_medio_min", ascending=False
).head(10)

top_10_piores_rotas
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    data=top_10_piores_rotas,
    x="atraso_medio_min",
    y=top_10_piores_rotas["origin"] + " ‚Üí " + top_10_piores_rotas["dest"]
)

plt.title("Top 10 Rotas com Maior Atraso M√©dio")
plt.xlabel("Atraso M√©dio (minutos)")
plt.ylabel("Rota")

plt.tight_layout()
plt.show()
```

```python
tabela_aeronaves = (
    df
    .groupby("tailnum")
    .agg(
        total_voos=("carrier", "count"),
        atraso_medio_min=("arr_delay", "mean")
    )
    .reset_index()
)

# Considerar apenas aeronaves com volume m√≠nimo
tabela_aeronaves = tabela_aeronaves[tabela_aeronaves["total_voos"] >= 50]

# Top 10 aeronaves com MAIOR atraso m√©dio
top_10_piores_aeronaves = tabela_aeronaves.sort_values(
    by="atraso_medio_min", ascending=False
).head(10)

top_10_piores_aeronaves
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    data=top_10_piores_aeronaves,
    x="atraso_medio_min",
    y="tailnum"
)

plt.title("Top 10 Aeronaves com Maior Atraso M√©dio")
plt.xlabel("Atraso M√©dio (minutos)")
plt.ylabel("Aeronave (Tail Number)")

plt.tight_layout()
plt.show()
```

```python
tabela_mes = (
    df
    .groupby("month")
    .agg(
        atraso_medio_min=("arr_delay", "mean"),
        percentual_atrasos=("atraso_chegada", "mean")
    )
    .reset_index()
)

tabela_mes["percentual_atrasos"] = tabela_mes["percentual_atrasos"] * 100

tabela_mes
```

```python
plt.figure(figsize=(8, 4))

sns.lineplot(
    data=tabela_mes,
    x="month",
    y="atraso_medio_min",
    marker="o"
)

plt.title("Atraso M√©dio por M√™s")
plt.xlabel("M√™s")
plt.ylabel("Atraso M√©dio (minutos)")
plt.xticks(range(1, 13))

plt.tight_layout()
plt.show()
```

```python
tabela_hora = (
    df
    .groupby("hour")
    .agg(
        atraso_medio_min=("arr_delay", "mean"),
        percentual_atrasos=("atraso_chegada", "mean")
    )
    .reset_index()
)

tabela_hora["percentual_atrasos"] = tabela_hora["percentual_atrasos"] * 100

tabela_hora
```

```python
plt.figure(figsize=(9, 4))

sns.lineplot(
    data=tabela_hora,
    x="hour",
    y="atraso_medio_min",
    marker="o"
)

plt.title("Atraso M√©dio por Hora do Dia")
plt.xlabel("Hora da Partida")
plt.ylabel("Atraso M√©dio (minutos)")

plt.tight_layout()
plt.show()
```

```python
df["dia_semana"] = pd.to_datetime(
    df[["year", "month", "day"]]
).dt.day_name()
```

```python
tabela_dia_semana = (
    df
    .groupby("dia_semana")
    .agg(
        atraso_medio_min=("arr_delay", "mean"),
        percentual_atrasos=("atraso_chegada", "mean")
    )
    .reset_index()
)

tabela_dia_semana["percentual_atrasos"] = tabela_dia_semana["percentual_atrasos"] * 100

tabela_dia_semana
```

```python
plt.figure(figsize=(9, 4))

sns.barplot(
    data=tabela_dia_semana,
    x="dia_semana",
    y="atraso_medio_min"
)

plt.title("Atraso M√©dio por Dia da Semana")
plt.xlabel("Dia da Semana")
plt.ylabel("Atraso M√©dio (minutos)")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Tabela de atraso m√©dio por hora
atraso_por_hora = (
    df.groupby("hour")["arr_delay"]
      .mean()
      .reset_index()
      .rename(columns={"hour": "hora", "arr_delay": "atraso_medio_min"})
)

# Gr√°fico + salvar imagem
plt.figure(figsize=(10, 5))
sns.lineplot(data=atraso_por_hora, x="hora", y="atraso_medio_min", marker="o")

plt.title("Atraso M√©dio por Hora do Dia")
plt.xlabel("Hora da Partida")
plt.ylabel("Atraso M√©dio (minutos)")

plt.tight_layout()
plt.savefig("images/atraso_por_hora.png", dpi=150, bbox_inches="tight")
plt.close()
```

```python
# Cria coluna de data e dia da semana (se ainda n√£o existir)
df["data"] = pd.to_datetime(df[["year", "month", "day"]])
df["dia_semana"] = df["data"].dt.day_name()  # Monday, Tuesday...

# Opcional: ordenar os dias na sequ√™ncia correta
ordem_dias = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df["dia_semana"] = pd.Categorical(df["dia_semana"], categories=ordem_dias, ordered=True)

# Tabela de atraso m√©dio por dia da semana
atraso_por_dia_semana = (
    df.groupby("dia_semana")["arr_delay"]
      .mean()
      .reset_index()
      .rename(columns={"arr_delay": "atraso_medio_min"})
)

# Gr√°fico + salvar imagem
plt.figure(figsize=(10, 5))
sns.barplot(data=atraso_por_dia_semana, x="dia_semana", y="atraso_medio_min")

plt.title("Atraso M√©dio por Dia da Semana")
plt.xlabel("Dia da Semana")
plt.ylabel("Atraso M√©dio (minutos)")
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("images/atraso_por_dia_semana.png", dpi=150, bbox_inches="tight")
plt.close()
```
